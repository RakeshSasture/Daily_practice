2. Write a Python program to read first n lines of a file.


```python
n = 3  # number of lines to read

with open("sample1.txt", "r") as file:
    for i in range(n):
        line = file.readline() # readline() -> reads one line at a time
        if not line:
            break   # stop if file ends before n lines
        print(line.strip()) # strip() -> removes extra newline spaces
```

4. Write a Python program to read last n lines of a file.


```python
n = 3  # number of last lines

with open("sample1.txt", "r") as file:
    lines = file.readlines()
    for line in lines[-n:]:
        print(line.strip())
```

7. Write a python program to find the longest words.


```python
with open("sample1.txt", "r") as file:
    words = file.read().split() # read().split() -> reads file and splits into words

# Find maximum length
max_length = max(len(word) for word in words) # max() -> finds length of longest word

# Find all longest words
longest_words = [word for word in words if len(word) == max_length]

print("Longest word(s):", longest_words)
print("Length:", max_length)
```

9. Write a Python program to count the frequency of words in a file.


```python
# Using Counter()-> counts frequency automatically
from collections import Counter
import string

with open("sample1.txt", "r") as file:
    text = file.read().lower() # lower() -> makes counting case-insensitive

# Remove punctuation
for ch in string.punctuation: # string.punctuation -> removes symbols like .,!?
    text = text.replace(ch, "")

words = text.split() # split() -> converts text into words

word_count = Counter(words)

print(word_count)
```

11. Write a Python program to write a list to a file.


```python
data = ["Python", "Java", "C++", "JavaScript"]

with open("sample1.txt", "w") as file: # w -> "w" mode -> creates file (or overwrites if exists)
    for item in data:
        file.write(item + "\n")

print("List written to file successfully.")
```

```python
# Using writelines()
data = ["Python\n", "Java\n", "C++\n", "JavaScript\n"]

with open("sample.txt", "w") as file:
    file.writelines(data) # writelines -> write list directly
```

13. Write a Python program to read a random line from a file.

15.Write a Python program that takes a text file as input and returns the number of words of a given text file.

```python
with open("sample.txt", "r") as file:
    text = file.read() # read() -> reads entire file as a string

words = text.split() # split() -> splits text into words (by spaces/newlines)

print("Number of words:", len(words)) # len() -> counts total words
```

```python
# Used line-by-line for large file
count = 0

with open("sample.txt", "r") as file:
    for line in file:
        words = line.split()
        count += len(words) # Counts words in each line separately

print("Number of words:", count)
```
